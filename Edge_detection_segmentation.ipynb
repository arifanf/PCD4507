```python
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
import cv2
from scipy import ndimage
```

SOBEL

```python
image = plt.imread('pemandangan_resize.jpg')
image.shape
plt.imshow(image)
```

```python
gray = rgb2gray(image)

horizontal = [[1,2,1],
              [0,0,0],
              [-1,-2,-1]]
vertical = [[-1,0,1],
            [-2,0,1],
            [-1,0,1]]
```

```python
out_h = ndimage.convolve(gray, horizontal, mode='reflect')
out_v = ndimage.convolve(gray, vertical, mode='reflect')
```

```python
plt.imshow(out_h, cmap="gray")
```

```python
plt.imshow(out_v, cmap='gray')
```

LAPLACIAN ZEROCROSSING

```python
laplace = [[-1,-1,-1],
           [-1,8,-1],
           [-1,-1,-1]]
```

```python
out_l = ndimage.convolve(gray, laplace, mode='reflect')
plt.imshow(out_l, cmap='gray')
```

CANNY

```python
edges = cv2.Canny(image,100,200)
plt.imshow(edges, cmap="gray")
```

PREWITT

```python
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(gray, -1, kernelx)
img_prewitty = cv2.filter2D(gray, -1, kernely)
```

```python
plt.imshow(img_prewittx, cmap='gray')
```

```python
plt.imshow(img_prewitty, cmap='gray')
```

ROBERT

```python
kernelx = np.array([[1,0],[0,-1]])
kernely = np.array([[0,1],[-1,0]])
img_robertx = cv2.filter2D(gray, -1, kernelx)
img_roberty = cv2.filter2D(gray, -1, kernely)
```

```python
plt.imshow(img_robertx, cmap='gray')
```

```python
plt.imshow(img_roberty, cmap='gray')
```
